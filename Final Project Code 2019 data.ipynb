```python
# Import Libraries.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Removing columns last_review and reviews_per_month
df = df.drop("last_review", axis=1)
```

```python
df = df.drop("reviews_per_month", axis=1)
```

```python
df.isna().sum()
```

```python
# All other nulls will be dropped with row.
df.dropna(inplace=True)
```

```python
df.isna().sum()
```

```python
# Creates a estimates revenue column
df["estimated_revenue"] = df["price"] * (df["availability_365"] / 365)
```

```python
# Creates a revenue by neighborhood average.
rev_by_neighborhood = df.groupby("neighbourhood_group")["estimated_revenue"].mean().sort_values(ascending=False)
```

```python
# Creates a revenue by room type.
rev_by_room = df.groupby("room_type")["estimated_revenue"].mean().sort_values(ascending=False)
```

```python
# Checks that all changes have been made.
df.info()
```

```python
plt.figure(figsize=(7,4))
rev_by_neighborhood.plot(kind="bar", color="#CC5500")
plt.title("Average Estimated Revenue by Borough 2019")
plt.ylabel("Average Revenue ($)")
plt.xlabel("Borough")
plt.xticks(rotation=45)
plt.grid(axis="y")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7,4))
rev_by_room.plot(kind="bar", color="#CC5500")
plt.title("Average Estimated Revenue by Room Type 2019")
plt.ylabel("Average Revenue ($)")
plt.xlabel("Room Type")
plt.xticks(rotation=45)
plt.grid(axis="y")
plt.tight_layout()
plt.show()
```

```python
# Creates a df for map data.
map_df = df[(df["price"] < 500)]
```

```python
# Scatter plot based on pricing.
plt.figure(figsize=(8, 6))
scatter = plt.scatter(
    map_df["longitude"],
    map_df["latitude"],
    c=map_df["price"],
    cmap="viridis",
    alpha=0.5,
    s=10
)

plt.colorbar(scatter, label="Price ($)")
plt.title("NYC Airbnb Listings by Price")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.grid(True)
plt.show()
```

```python
# Imports for sklearn.
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import mean_squared_error
```

```python
# Selecting features and targe from columns in our data base.
features = df[["room_type","neighbourhood_group","minimum_nights", "number_of_reviews","availability_365"]]
target = df["price"]
```

```python
# Filtering for outliers.
df = df[(df["price"] > 10) & (df["price"] < 500)]
```

```python
# One hot encode. Turns data into binary for better analysis.
features_encoded = pd.get_dummies(features, drop_first=True)
```

```python
# Train and test split.
x_train, x_test, y_train, y_test = train_test_split(features_encoded, target, test_size=0.2, random_state=42)
```

```python
# Train model.
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
# Preditctions.
y_pred = model.predict(x_test)
```

```python
# Evaluation.
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error" + " " + str(mse))
```
