```python
# Removing Hotel Room type from data set as the 2019 version does not have it.
initial_count = len(df)
df = df[df["room_type"] != "Hotel room"]
final_count = len(df)
removed_count = initial_count - final_count
print(removed_count)
```

```python
# Creates a estimates revenue column
df["estimated_revenue"] = df["price"] * (df["availability_365"] / 365)
```

```python
# Creates a revenue by neighborhood average.
rev_by_neighborhood = df.groupby("neighbourhood_group")["estimated_revenue"].mean().sort_values(ascending=False)
```

```python
# Creates a revenue by room type.
rev_by_room = df.groupby("room_type")["estimated_revenue"].mean().sort_values(ascending=False)
```

```python
plt.figure(figsize=(7,4))
rev_by_neighborhood.plot(kind="bar", color="#CC5500")
plt.title("Average Estimated Revenue by Borough 2024")
plt.ylabel("Average Revenue ($)")
plt.xlabel("Borough")
plt.xticks(rotation=45)
plt.grid(axis="y")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7,4))
rev_by_room.plot(kind="bar", color="#CC5500")
plt.title("Average Estimated Revenue by Room Type 2024")
plt.ylabel("Average Revenue ($)")
plt.xlabel("Room Type")
plt.xticks(rotation=45)
plt.grid(axis="y")
plt.tight_layout()
plt.show()
```

```python
# Creates a df for map data.
map_df = df[(df["price"] < 500)]
```

```python
# Scatter plot based on pricing.
plt.figure(figsize=(8, 6))
scatter = plt.scatter(
    map_df["longitude"],
    map_df["latitude"],
    c=map_df["price"],
    cmap="viridis",
    alpha=0.5,
    s=10
)

plt.colorbar(scatter, label="Price ($)")
plt.title("NYC Airbnb Listings by Price")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.grid(True)
plt.show()
```
